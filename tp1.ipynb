```python
import numpy as np
from sklearn import cluster
```

we use Kmeans algorithm to group the dataset into clusters

```python
#n_clusters paramither is used to determinate the number of clusters 
#in this case we used clusters since the CAH give us the result of 4 logic clusters
#also fit() is a function that start the train , it take the dataset as a paramither

kmeans = cluster.KMeans(n_clusters=4)
kmeans.fit(fromage)

```

```python
np.random.seed(0)

idk = np.argsort(kmeans.labels_)
print(pd.DataFrame(fromage.index[idk],kmeans.labels_[idk]))
print(kmeans.transform(fromage))

#after training and saving the model into kmeans variable, we print the result each "fromage name or category" with their group index (0..3)
```

```python
#print(pd.DataFrame(fromage.index[idk],kmeans.labels_[idk]))
kmeans.cluster_centers_
```

```python
from sklearn import metrics
#utilisation de la mtrique "silhouette"
res = np.arange(9, dtype ="double")
for k in np.arange(9):
 km = cluster.KMeans(n_clusters = k+2)
 km.fit(fromage)
 res[k] = metrics.silhouette_score(fromage,km.labels_)
print (res)
#graphique
import matplotlib.pyplot as plt
```

```python
plt.title("silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,11,1),(res))
plt.show
```

```python
import pandas
import numpy as np
#librairies pour la CAH
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
Z = linkage(fromage,method='ward', metric='euclidean')
#affichage du dendrogramme
plt.title("CAH")
plt.title('CAH avec matrialisation des 4 classes')
dendrogram(Z,labels=fromage.index,orientation='left',color_threshold=255)
plt.show()
groupes_cah = fcluster(Z, t = 255,criterion='distance')
print(groupes_cah)
#index tris des groupes
idg = np.argsort(groupes_cah)
#affichage des observations et leurs groupes
print(pandas.DataFrame(fromage.index[idg],groupes_cah[idg]))
```
