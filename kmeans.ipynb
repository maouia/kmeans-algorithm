1   Initialize random centroids
2   Label each data point
3   Update centroids 
4   Repeat steps 3 or 4 until centoids stops changing

```python
def random_centroids(data,k):
    centroids = []
    for i in range(k):
        centroid = data.apply(lambda x:float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids,axis=1)
```

```python
centroids = random_centroids(data,4)
```

```python
centroids
```

```python
#calculate the distence each column to to centroid 
#and square each diffrence together 
#take distance formula



np.sqrt(((data - centroids.iloc[:,0])**2).sum(axis=1))
```

```python
#and apply that to all centraids
distances = centroids.apply(lambda x : np.sqrt(((data - x)**2).sum(axis=1)))
```

```python
#make it in a function
def get_labels(data,centroids):
    distances = centroids.apply(lambda x : np.sqrt(((data - x)**2).sum(axis=1)))
    return distances.idxmin(axis='columns')
```

```python
labels = get_labels(data,centroids)
```

```python
labels.value_counts()
```

```python
#defind th geometric mean
def new_centroids(data,labels,k):
    return data.groupby(labels).apply(lambda x : np.exp(np.log(x).mean())).T
```

Data visualisation

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data,labels,centroids,iteration):
    pca = PCA()
    data_2D = pca.fit_transform(data)
    centroids_2D = pca.fit_transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2D[:,0],y=data_2D[:,1],c=labels)
    plt.scatter(x=centroids_2D[:,0],y=centroids_2D[:,1])
    plt.show()
```

```python
max_iteration = 100
k=4

centroids = random_centroids(data,k)
old_centroids = pd.DataFrame()
iteration=1
while iteration < max_iteration and not centroids.equals(old_centroids):
    old_centroids = centroids

    labels = get_labels(data,centroids)
    centroids = new_centroids(data,labels,k)
    plot_clusters(data,labels,centroids,iteration)
    iteration +=1
```
